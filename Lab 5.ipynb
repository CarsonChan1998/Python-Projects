```r
data = read.csv(file="SAT_Participation_Demo.csv", header=TRUE, sep=",")
col4 = data[,4]
col6 = data[,6]
mean4 = mean(col4)
mean6 = mean(col6)
combined = c()
if (mean4 < mean6){
    combined = c(col4, col6 + 175)
    colnames = c((colnames(data))[4], (colnames(data))[6])
} else {
    combined = c(col6 + 175, col4)
    colnames = c( (colnames(data))[6], (colnames(data))[4])

}


matrix = matrix(combined, ncol = 2)
colnames(matrix) = colnames
boxplot(matrix, main = "Ordered Box Plot")
```
