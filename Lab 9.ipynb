```python
import pandas
import matplotlib.pyplot as plt
import numpy as np

from sklearn.cluster import KMeans

data = pandas.read_csv("features.csv")

plt.scatter(data.iloc[:,0],data.iloc[:,1])
plt.show()
print("There are 3 main groups")
```

```python
k = [1,2,3,4,5]
# fig, axs = plt.subplots(2, 3)



fig = plt.figure()
fig.set_figheight(15)
fig.set_figwidth(20)

fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 6):
    ax = fig.add_subplot(2, 3, i)
    ax.set_title("Number of clusters: "+ str(i))
    ax.set(xlabel='V1', ylabel='V2')
    
    
    kmeans = KMeans(n_clusters=i).fit(data.values)
    y_kmeans = kmeans.predict(data.values)
    ax.scatter(data.values[:,0], data.values[:,1], c = y_kmeans, s=25, cmap = "viridis")
    centers = kmeans.cluster_centers_
    ax.scatter(centers[:, 0], centers[:, 1], c='cyan', s=50, alpha=1)
centers
```
