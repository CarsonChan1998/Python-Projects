```r
library(ggplot2)
data = read.csv("cars_information")
par(mfrow=c(2,2))
dataSubset = data.frame(data$mpg, data$disp, data$drat, data$wt)

dataCor = data.frame(cor(dataSubset, dataSubset))

dataCor$data.mpg[1] = 0
dataCor$data.disp[2]= 0
dataCor$data.drat[3] = 0
dataCor$data.wt[4] = 0
# max(dataCor)
# min(dataCor)
max1 = c()
max2 = c()
min1 = c()
min2 = c()


for (i in c(1:4)){
    if (max(dataCor) %in% dataCor[,i]){
            max1 = colnames(dataCor)[i]
            max2 = rownames(dataCor)[match(max(dataCor),dataCor[,i])]
        break
        
    }  
}

for (i in c(1:4)){
    if (min(dataCor) %in% dataCor[,i]){
            min1 = colnames(dataCor)[i]
            min2 = rownames(dataCor)[match(min(dataCor),dataCor[,i])]
        break
        
    }  
}

maxDataFrame = data.frame(xaxis = c(dataSubset[,max1]), yaxis = c(dataSubset[,max2]))
minDataFrame = data.frame(xaxis = c(dataSubset[,min1]), yaxis = c(dataSubset[,min2]))
ggplot(data = maxDataFrame, aes(x = xaxis, y= yaxis, group = 1)) + geom_point() + labs(title="Plot 1" ,x = max1, y=max2)+
geom_smooth(method ="loess")
ggplot(data = minDataFrame, aes(x = xaxis, y= yaxis, group = 1)) + geom_point() + labs(title="Plot 2" ,x = min1, y=min2)+
geom_smooth(method ="loess")
# plot(dataSubset[,max1],dataSubset[,max2], xlab = max1, ylab =  max2, main = "Plot 1")

# plot(dataSubset[,min1],dataSubset[,min2], xlab = min1, ylab =  min2, main = "Plot 2")

```

```r

library(ggplot2)
data = read.csv("Number_of_Crimes")


dataFrame = data.frame(period =c("1996-1997", "2003-2004"), ncrimes = data$ncrimes)

ggplot(data= dataFrame , aes(x=period, y = ncrimes)) + ylim(38000,48000)+ geom_bar(stat="identity") + 
geom_segment(aes(x = period[1], y = ncrimes[1], xend = period[2], yend = ncrimes[2], colour = "Line"), data = dataFrame)

ggplot(data= dataFrame , aes(x=period, y = ncrimes)) + ylim(0,48000)+ geom_bar(stat="identity") + 
geom_segment(aes(x = period[1], y = ncrimes[1], xend = period[2], yend = ncrimes[2], colour = "Line"), data = dataFrame)


# barplot(data$ncrimes, ylim=c(38000,48000))
# lines(data$ncrimes)
# barplot(data$ncrimes, ylim=c(0,48000))
# lines(data$ncrimes)


#Line gradient is steeper in 1 compared to 2
```
