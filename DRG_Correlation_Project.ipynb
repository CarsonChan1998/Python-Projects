```python
import statistics
import pandas
import heapq 
import numpy
from scipy import stats
data = pandas.read_csv("IPPS_DRG_FY2017.csv")

Question1_1 = []

#a)#
answer = []
provider_id = list(data.iloc[:,1])
provider_state = list(data.iloc[:,5])

provider_id_unique = sorted(list(set(data.iloc[:,1])))
provider_state_unique = sorted(set(provider_state))

providerIdProviderState = (pandas.DataFrame(list(zip(provider_id, provider_state)), 
               columns =['Provider Id', 'Provider State'])).drop_duplicates()
providerIdProviderState= providerIdProviderState.reset_index(drop=True)


count = []
state = []
for i in provider_state_unique:
    temp = data.loc[data["Provider State"] == i]
    
    count.append(len(list(set(temp.iloc[:,1]))))
    state.append(i)
    
templist = [count, state]
most_unique_provider_state = templist[1][templist[0].index(max(templist[0]))]

answer.append(len(provider_id_unique))
answer.append(most_unique_provider_state)
Question1_1.append(answer)
# print(answer)



#b)#
answer = []
hospital_discharges = list(data.iloc[:,8])
mean_hospital_discharges = statistics.mean(hospital_discharges)
median_hospital_discharges = statistics.median(hospital_discharges)
std_hospital_discharges = statistics.stdev(hospital_discharges)
answer.append(mean_hospital_discharges)
answer.append(median_hospital_discharges)
answer.append(std_hospital_discharges)
Question1_1.append(answer)
# print(answer)
#c)#
answer = []
drg_definition = list(data.iloc[:,0])
drg_definition_unique = sorted(list(set(drg_definition)))
mode_drg_definition_unique = statistics.mode(drg_definition)
answer.append(len(drg_definition_unique))
answer.append(mode_drg_definition_unique)
Question1_1.append(answer)
# print(answer)

#d)#
answer = []
DRG_Categories = []
for i in drg_definition_unique:
    HD = []
    temp = data.loc[data["DRG Definition"] == i]
    HD = len(temp.iloc[:,8])
    DRG_Categories.append(HD)
    
least_hospital_discharges = drg_definition_unique[DRG_Categories.index(min(DRG_Categories))]
length = len(data.loc[data["DRG Definition"] == least_hospital_discharges])
answer.append(least_hospital_discharges)
answer.append(length)
Question1_1.append(answer)
print(answer)


# Question1_1
```

```python
#most difference = AR vs CA
#H0 = There is no difference in the means of AR and CA DRG charge 066.
#H1 = The mean for CA DRG charge 066 is more than the mean for AR DRG charge 066. 

# Scipy always gives the test statistic as signed. 
# This means that given p and t values from a two-tailed test, 
# you would reject the null hypothesis of a greater-than test when p/2 < alpha and t > 0, 
# and of a less-than test when p/2 < alpha and t < 0.

alpha = 0.1
AR = States[2].iloc[:,2].dropna()
CA = States[4].iloc[:,2].dropna()

result = stats.ttest_ind(CA, AR)
testStat = result[0]
pValue = result[1]
pValueForOneSided = pValue/2
pValueForOneSided
print("Since the test statistic is "+str(testStat)+" > 0 and the pValue for one sided t test is "+str(pValueForOneSided)+
      " < alpha = 0.1, thus, we reject the null hypothesis and conclude that the mean DRG charge of CA is indeed more " +
      "than the mean DRG charge of AR.")
```
