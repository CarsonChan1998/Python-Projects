```r
library(ggplot2)
dataframe = data.frame( xaxis = input$Age ,yaxis = input$BloodPressure)
# dataframe
ggplot(data = dataframe, aes(x = xaxis, y= yaxis, group = 1)) + geom_point() +
labs(title="Relation between Blood-Pressure and Age" ,x = "Age", y="Blood Pressure")+
geom_hline(data=dataframe, aes(yintercept=mean(input$BloodPressure)),linetype="dashed", size=1, colour="red")+
geom_smooth(method ="loess")
```

```r
library(ggplot2)
quart_1 <- subset(input, input$Aneurisms_q1 <= quantile(input$Aneurisms_q1, 0.25))
quart_2 <- subset(input, input$Aneurisms_q1 > quantile(input$Aneurisms_q1, 0.25) & input$Aneurisms_q1 <= quantile(input$Aneurisms_q1, 0.50))
quart_3 <- subset(input, input$Aneurisms_q1 > quantile(input$Aneurisms_q1, 0.50) & input$Aneurisms_q1 <= quantile(input$Aneurisms_q1, 0.75))
quart_4 <- subset(input, input$Aneurisms_q1 > quantile(input$Aneurisms_q1, 0.75))

all = data.frame(quart1 = quart_1$Age,quart2 = quart_2$Age,quart3 = quart_3$Age,quart4 = quart_4$Age)
df = stack(all)
names(df) <- c("Ages", "Quarts")
ggplot(df, aes(x=Ages)) + geom_density(alpha=0.25,aes(group=Quarts, colour=Quarts, fill=Quarts))+ labs(title = "Age Density Plot for Aneurysms_q1")+
geom_vline(data=subset(df,Quarts == 'quart1'), aes(xintercept=mean(quart_1$Age)),linetype="dashed", size=1, colour="red")+
geom_vline(data=subset(df,Quarts == 'quart2'), aes(xintercept=mean(quart_2$Age)),linetype="dashed", size=1, colour="green")+
geom_vline(data=subset(df,Quarts == 'quart3'), aes(xintercept=mean(quart_3$Age)),linetype="dashed", size=1, colour="cyan")+
geom_vline(data=subset(df,Quarts == 'quart4'), aes(xintercept=mean(quart_4$Age)),linetype="dashed", size=1, colour="purple")+
geom_density(aes(df$Ages), size = 1)+ xlim(7.5,25)
```

```r
input = read.csv("aneurysm_data.csv", header = TRUE, sep = ",")
dataframe = data.frame(xx = c(input$Aneurisms_q3, input$Aneurisms_q4), Group = rep(c("q3","q4"),each = 100) )
ggplot(dataframe,aes(x=xx, fill = Group)) + 
    geom_histogram(aes(y= stat(density)), data=subset(dataframe,Group == 'q3'), alpha = 0.5, binwidth = 10) + 
    geom_histogram(aes(y= stat(density)),data=subset(dataframe,Group == 'q4'), alpha = 0.5, binwidth = 10) + labs(title = "Histogram of Aneurisms Q3 vs Q4")+xlab("Number of Aneurisms")+
    geom_vline(data=subset(dataframe,Group == 'q3'), aes(xintercept=mean(input$Aneurisms_q3)),linetype="dashed", size=1, colour="red")+
    geom_vline(data=subset(dataframe,Group == 'q4'), aes(xintercept=mean(input$Aneurisms_q4)),linetype="dashed", size=1, colour="blue")+
  stat_function(
    fun = dnorm, 
    args = list(mean = mean(input$Aneurisms_q3), sd = sd(input$Aneurisms_q3)), 
    lwd = 2, 
    col = 'red'
  )+
stat_function(
    fun = dnorm, 
    args = list(mean = mean(input$Aneurisms_q4), sd = sd(input$Aneurisms_q4)), 
    lwd = 2, 
    col = 'blue'
  )
```
