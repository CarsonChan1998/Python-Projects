```python
import pandas
data = pandas.read_csv("ratings.csv")
data
```

```python
movieUnique = sorted((set(data.iloc[:,1])))
userUnique = sorted((set(data.iloc[:,0])))
newDF = pandas.DataFrame(movieUnique)  
newDF.columns = ["movieId"]

for i in userUnique:
    temp = data.loc[data["userId"] == i].iloc[:,1:3]
    temp.columns = ["movieId", i]
# newDF
    newDF = pandas.merge(newDF, temp, how='outer', on="movieId")

newDF.fillna(0, inplace=True)
newDF = newDF.T
newDF.columns = movieUnique
newDF.index.names = ["User ID"]
newDF = newDF.drop("movieId")
newDF
```

```python
newDF = newDF.T
```

```python
import sklearn
import numpy
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn import preprocessing 
import statistics

matrix = newDF.values
newDF_scaled = sklearn.preprocessing.scale(matrix, with_std= False)
newDF_scaled
```

```python
pca = PCA(n_components=2)
# pca.fit(newDF_scaled)
newDF_trans = pca.fit_transform(newDF_scaled)
# pca.fit(newDF_trans)

plt.scatter(newDF_trans[:,0],newDF_trans[:,1])
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()

```

```python

plt.scatter([1,1], pca.explained_variance_ratio_)
sum(pca.explained_variance_ratio_)
# diff = pca.explained_variance_ratio_[0]-pca.explained_variance_ratio_[1]
# print("Difference in variance: "+ str(diff))
```

```python
from sklearn.cluster import KMeans

k = [2,4,8,16,32]
inertia = []
for i in k:
    print("Processing: "+ str(i))
    kmeans = KMeans(n_clusters=i, random_state=1)
    result = kmeans.fit(matrix)
    inertia.append(result.inertia_)
    
    y_kmeans = kmeans.predict(matrix)
print("Done")



```

```python
print(inertia)

plt.plot(k, inertia, marker='o')
plt.title("Elbow plot")
plt.xlabel("N clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
# u, s, vt = numpy.linalg.svd(matrix)
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=32, random_state= 5)
svd.fit(matrix)
svd.singular_values_
plt.plot(svd.singular_values_)
```

```python
k = [2,4,8,16,32]

result = []
for i in k:
    svd = TruncatedSVD(n_components=i, random_state= 5)
    temp = svd.fit(matrix)
    result.append(temp.explained_variance_ratio_.sum())

plt.plot(k, result, marker='o')
result
#compared to the inertia graph, it is vertically inverted.
```
